## The dataset

```python
import pandas as pd

data = pd.read_json('https://raw.githubusercontent.com/aurioldegbelo/dgp2021-datasets/main/weather_stations.json')
data
```

## Diplaying the attributes

```python
import altair as alt

w = 500
h = 500

station_elevation = alt.Chart(data).mark_tick().encode(
    y ='elevation:Q',
    x ='station:N'
).properties(width=w, height=h)
station_elevation
```

## Displaying the boundaries

```python
# https://altair-viz.github.io/gallery/point_map.html
url_nrw = "https://raw.githubusercontent.com/aurioldegbelo/dgp2021-datasets/main/geo/boundaries-nrw.geojson"
bd_nrw = alt.Data(url=url_nrw, format=alt.DataFormat(property="features"))
state = alt.Chart(bd_nrw).mark_geoshape(tooltip="North Rhine-Westphalia", fill ="#eee",  stroke = "#757575",  strokeWidth= 0.5)
state
```

```python
url_bochum = "https://raw.githubusercontent.com/aurioldegbelo/dgp2021-datasets/main/geo/boundaries-bochum.geojson"
bd_bochum = alt.Data(url=url_bochum, format=alt.DataFormat(property="features"))
city = alt.Chart(bd_bochum).mark_geoshape(tooltip="Bochum", fill ="#eee",  stroke = "#757575",  strokeWidth= 0.5)
city
```

```python
(state + city)#.properties(width=w, height=h)
```

boundaries + stations (their geographic location in NRW)

```python
stations = alt.Chart(data).mark_point( # mark_circle
    size=20,
    color='steelblue'
).encode(
    longitude='lon:Q',
    latitude='lat:Q',
    color = 'station:N',
    tooltip=['name']
)
state + city + stations
```

boundaries + location of stations (as shape)

```python
station_location = alt.Chart(data).mark_point().encode(
    longitude='lon:Q',
    latitude ='lat:Q',
    shape = 'name:N',
    tooltip=['name']
).properties(width=w, height=h)
state + city + station_location
```

boundaries + types of stations

```python
station_type = alt.Chart(data).mark_circle(
    size=20,
    color='steelblue'
).encode(
    longitude='lon:Q',
    latitude='lat:Q',
    color = 'station_type:N',
    tooltip=['name']
)
state + city + station_type
```

Elevation of the stations (proportional symbol map)

```python
elevation = alt.Chart(data).mark_circle().encode(
    longitude='lon:Q',
    latitude='lat:Q',
    size = 'elevation:Q',
    tooltip=['name']
).properties(width=w, height=h)
state + city + elevation
```

Elevation and type of station (dots as circles)

```python
elevation_per_station_type = alt.Chart(data).mark_circle().encode(
    longitude='lon:Q',
    latitude='lat:Q',
    size = 'elevation:Q',
    color = 'station_type:N',
    tooltip=['name']
).properties(width=w, height=h)
state + city + elevation_per_station_type
```

Average temperature value per station (dots as circles)


```python
temp_per_station_type = alt.Chart(data).mark_circle().encode(
    longitude='lon:Q',
    latitude='lat:Q',
    size = 'avgTemperature:Q',
    color = 'name:N',
    tooltip=['name']
).properties(width=w, height=h)
state + city + temp_per_station_type
```
