## Displaying the boundaries

```python
# https://altair-viz.github.io/gallery/point_map.html
url_nrw = "https://raw.githubusercontent.com/aurioldegbelo/hsb-dgp-datasets/main/geo/boundaries-nrw.geojson"
bd_nrw = alt.Data(url=url_nrw, format=alt.DataFormat(property="features"))

state = alt.Chart(bd_nrw).mark_geoshape(tooltip="North Rhine-Westphalia", fill ="#eee",  stroke = "#101010",  strokeWidth= 0.5)
state
```

```python
url_bochum = "https://raw.githubusercontent.com/aurioldegbelo/hsb-dgp-datasets/main/geo/boundaries-bochum.geojson"
bd_bochum = alt.Data(url=url_bochum, format=alt.DataFormat(property="features"))
city = alt.Chart(bd_bochum).mark_geoshape(tooltip="Bochum", fill ="#eee",  stroke = "#101010",  strokeWidth= 0.5)
city
```

```python
import altair as alt
import pandas as pd

# Example data: weather stations with latitude, longitude, and station name
data = {
    'station_name': ['Station A', 'Station B', 'Station C', 'Station D'],
    'latitude': [34.0522, 36.7783, 40.7128, 51.5074],
    'longitude': [-118.2437, -119.4179, -74.0060, -0.1278]
}

# Convert data into a pandas DataFrame
df = pd.DataFrame(data)

# Melt the DataFrame to long format for plotting both latitude and longitude
df_melted = df.melt(id_vars=['station_name'], value_vars=['latitude', 'longitude'],
                    var_name='coordinate_type', value_name='value')

# Create the bar chart with Altair
chart = alt.Chart(df_melted).mark_bar().encode(
    y='station_name:N',  # Categorical axis for stations
    x='value:Q',  # Quantitative axis for latitude/longitude values
    color='coordinate_type:N',  # Different colors for latitude and longitude
    tooltip=['station_name', 'coordinate_type', 'value']
).properties(
    title='Latitude and Longitude of Weather Stations'
)

# Display the chart
chart.show()
```
