# Demo 2: Visualizing Air Quality Data

## The Dataset

```python
# Data from the LANUV https://www.lanuv.nrw.de/umwelt/luft/immissionen/aktuelle-luftqualitaet
# Messungen der grorumigen Luftschadstoffbelastung (Hintergrund), verkehrsnahe Messungen sowie Messungen an Stationen in der Nhe von Industriebetrieben bilden das Basismessnetz
# https://www.lanuv.nrw.de/fileadmin/lanuvpubl/1_infoblaetter/LANUV_Info65_Luft%C3%BCberwachung.pdf

import pandas as pd

data = pd.read_json('https://raw.githubusercontent.com/aurioldegbelo/hsb-dgp-datasets/main/weather_stations.json')

data
```

## Diplaying the attributes

```python
import altair as alt

w = 400
h = 400

alt.Chart(data).mark_bar().encode(
    y ='name:N',
    x ='elevation:Q',
).properties(width=w, height=h)
```

```python
w = 400
h = 400

t = alt.Chart(data).mark_bar(color='red').encode(
    y ='name:N',
    x ='avgTemperature:Q',
).properties(width=w, height=h)
t
```

## Displaying the boundaries

```python
# https://altair-viz.github.io/gallery/point_map.html
url_nrw = "https://raw.githubusercontent.com/aurioldegbelo/hsb-dgp-datasets/main/geo/boundaries-nrw.geojson"
bd_nrw = alt.Data(url=url_nrw, format=alt.DataFormat(property="features"))

state = alt.Chart(bd_nrw).mark_geoshape(tooltip="North Rhine-Westphalia", fill ="#eee",  stroke = "#101010",  strokeWidth= 0.5)
state
```

```python
url_bochum = "https://raw.githubusercontent.com/aurioldegbelo/hsb-dgp-datasets/main/geo/boundaries-bochum.geojson"
bd_bochum = alt.Data(url=url_bochum, format=alt.DataFormat(property="features"))
city = alt.Chart(bd_bochum).mark_geoshape(tooltip="Bochum", fill ="#eee",  stroke = "#101010",  strokeWidth= 0.5)
city
```

```python
(state + city).properties(width=w, height=h)
```

boundaries + stations (their geographic location in NRW)

```python
stations = alt.Chart(data).mark_square().encode(
    longitude='lon:Q',
    latitude='lat:Q',
    color = 'station:N',
    tooltip=['name']
)
(state + city + stations).properties(width=w, height=h)
```

boundaries + location of stations (as shape)

```python
station_location = alt.Chart(data).mark_point().encode(
    longitude='lon:Q',
    latitude ='lat:Q',
    shape = 'name:N',
    color = 'name:N',
    tooltip=['name']
).properties(width=w, height=h)
state + city + station_location
```

boundaries + types of stations

```python
w = 500
h = 500
station_type = alt.Chart(data).mark_circle().encode(
    longitude='lon:Q',
    latitude='lat:Q',
    color = 'station_type:N',
    tooltip=['name']
)
(state + city + station_type).properties(width=w, height=h)
```

Elevation of the stations (proportional symbol map - 1 variable)

```python
elevation = alt.Chart(data).mark_circle().encode(
    longitude='lon:Q',
    latitude='lat:Q',
    size = 'elevation:Q',
    tooltip=['name']
).properties(width=w, height=h)
state + city + elevation
```

Elevation and type of station (dots as circles - 2 variables)

```python
elevation_per_station_type = alt.Chart(data).mark_circle().encode(
    longitude  ='lon:Q',
    latitude ='lat:Q',
    size = 'elevation:Q',
    color = 'station_type:N',
    tooltip=['name']
).properties(width=w, height=h)
state + city + elevation_per_station_type
```

Average temperature value per station (dots as circles - 2 variables)


```python
temp_per_station_type = alt.Chart(data).mark_circle().encode(
    longitude='lon:Q',
    latitude='lat:Q',
    size = 'avgTemperature:Q',
    color = 'name:N',
    tooltip=['name']
).properties(width=w, height=h)
state + city + temp_per_station_type
```

```python
import pandas as pd
import altair as alt

# Sample data: weather stations
data = pd.DataFrame({
    "station": [
        "Station A", "Station B", "Station C", "Station D",
        "Station E", "Station F"
    ],
    "elevation_m": [50, 200, 450, 800, 1200, 1800],
    "avg_temperature_c": [18.5, 16.2, 14.0, 11.5, 9.2, 6.0]
})

# Create a 2D scatter plot
chart = (
    alt.Chart(data)
    .mark_circle(size=120)
    .encode(
        x=alt.X(
            "elevation_m:Q",
            title="Elevation (meters)"
        ),
        y=alt.Y(
            "avg_temperature_c:Q",
            title="Average Temperature (C)"
        ),
        tooltip=[
            alt.Tooltip("station:N", title="Station"),
            alt.Tooltip("elevation_m:Q", title="Elevation (m)"),
            alt.Tooltip("avg_temperature_c:Q", title="Avg Temp (C)")
        ]
    )
    .properties(
        title="Elevation vs. Average Temperature of Weather Stations",
        width=600,
        height=400
    )
)

chart
```

```python
import pandas as pd
import altair as alt

# Sample data: weather stations with coordinates
data = pd.DataFrame({
    "station": [
        "Station A", "Station B", "Station C",
        "Station D", "Station E"
    ],
    "latitude": [52.52, 48.85, 40.71, 34.05, -33.87],
    "longitude": [13.40, 2.35, -74.01, -118.24, 151.21]
})

# Convert data to long format
data_long = data.melt(
    id_vars="station",
    value_vars=["latitude", "longitude"],
    var_name="coordinate_type",
    value_name="coordinate_value"
)

# Create a bar chart
chart = (
    alt.Chart(data_long)
    .mark_bar()
    .encode(
        y=alt.X(
            "station:N",
            title="Weather Station"
        ),
        x=alt.Y(
            "coordinate_value:Q",
            title="Coordinate Value"
        ),
        color=alt.Color(
            "coordinate_type:N",
            title="Coordinate Type"
        ),
        tooltip=[
            alt.Tooltip("station:N", title="Station"),
            alt.Tooltip("coordinate_type:N", title="Type"),
            alt.Tooltip("coordinate_value:Q", title="Value")
        ]
    )
    .properties(
        title="Latitude and Longitude of Weather Stations",
        width=600,
        height=400
    )
)

chart
```
